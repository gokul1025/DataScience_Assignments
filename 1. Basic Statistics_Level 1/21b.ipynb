Check Whether the Adipose Tissue (AT) and Waist Circumference(Waist)  from wc-at data set  follows Normal Distribution 
       Dataset: wc-at.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
q21b = pd.read_csv("wc-at.csv")
q21b
```

```python
plt.hist(q21b['Waist'], edgecolor= 'yellow')
plt.show()
```

```python
plt.boxplot(x = 'Waist', data= q21b)
plt.title("Waist")
#plt.savefig('Waist.png')
plt.show()
```

```python
sn.distplot(q21b['Waist'], 
             bins=10,
            kde = True
            )
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(q21b['Waist'])
plt.show()
```

```python
import scipy.stats as stats
stats.probplot(q21b['Waist'], dist = 'norm', plot = plt)
plt.xlabel('Waist')
plt.savefig('Waist.png')
plt.show()
```

```python
sn.distplot(q21b['AT'], bins =10, kde=True)
plt.show()
```

```python
sm.qqplot(q21b['AT'])
plt.show()
```

```python
stats.probplot(q21b['AT'], dist = 'norm', plot = plt)
plt.xlabel('AT')
plt.savefig('AT.png')
plt.show()
```
